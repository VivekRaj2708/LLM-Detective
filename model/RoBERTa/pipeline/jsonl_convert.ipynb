```python
import json

inp = "/home/harinarayan.j/LLM-DetectAIve/script/samples.json"
out = "/home/harinarayan.j/LLM-DetectAIve/script/samples.jsonl"

with open(inp, "r") as f:
    data = json.load(f)

with open(out, "w") as f:
    for item in data:
        f.write(json.dumps(item) + "\n")

print("✅ Converted to JSONL:", out)
```

```python
import json
import uuid

INPUT = "/home/harinarayan.j/LLM-DetectAIve/script/samples.jsonl"
OUTPUT = "/home/harinarayan.j/LLM-DetectAIve/script/samples_converted.jsonl"

with open(INPUT, "r") as fin, open(OUTPUT, "w") as fout:
    for line in fin:
        line = line.strip()
        if not line:
            continue

        arr = json.loads(line)
        if not isinstance(arr, list) or len(arr) != 2:
            print("❌ Skipping invalid line:", line)
            continue

        text, label = arr

        obj = {
            "id": str(uuid.uuid4()),   # or use existing if you have one
            "text": text,
            "label": label             # ✅ label stays the same
        }
        fout.write(json.dumps(obj) + "\n")

print("✅ Converted and saved to:", OUTPUT)
```
